```python
import pandas as pd
import datetime as dt
from re import sub
from decimal import Decimal
```

```python
load_df = pd.read_csv('output/complete_movie_list_df_2021_2022.csv')
load_df = load_df.drop(['Unnamed: 0.1','Production','Year'], axis=1)
```

```python
load_df
```

```python
load_df = load_df.fillna(0)
load_df
```

```python
#converting to proper types
load_df['Released'] = pd.to_datetime(load_df['Released'], errors='coerce')
load_df.Title = load_df.Title.astype('string')
load_df.Rated = load_df.Rated.astype('string')
load_df.Runtime = load_df.Runtime.astype('string')
#load_df['Runtime'] = load_df['Runtime'].str.split(' ', 1, expand=True)
load_df.Genre = load_df.Genre.astype('string')
load_df.Director = load_df.Director.astype('string')
load_df.Writer = load_df.Writer.astype('string')
load_df.Actors = load_df.Actors.astype('string')
load_df.Director = load_df.Director.astype('string')
load_df.Language = load_df.Language.astype('string')
load_df['BoxOffice'] = load_df['BoxOffice'].replace('[\$,]', '', regex=True).astype(int)
load_df.dtypes
```

```python
converted_df = load_df
converted_df
```

```python
#clean dataframe
converted_df = pd.DataFrame(converted_df.rename(columns={"Unnamed: 0":"movie_id", 'Title':'title','Rated':'movie_rating','Released':'release_date',
'Runtime':'runtime', 'Genre': 'genre_name','Director':'director','Writer':'writer','Actors':'Actors', 'Language':'language', 'Metascore':'meta_score','imdbRating':'imdb_rating'}))

#export to csv
converted_df.to_csv ('output/clean_movie_list_df.csv', header=True, index=False)
converted_df
```
