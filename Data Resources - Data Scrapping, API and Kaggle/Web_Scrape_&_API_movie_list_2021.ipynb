```python
# Dependencies
import requests
from bs4 import BeautifulSoup as bs
import requests
import numpy as np
import time
import pandas as pd
import csv
import json
```

```python
def scrape2021():
    pages = np.arange(1,400,100)
    years = np.arange(2021,2022,1)
    URL = 'https://www.the-numbers.com/box-office-records/domestic/all-movies/cumulative/released-in-'
    HEADERS = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
    movies = []

    for year in years:
        #year = requests.get('https://www.the-numbers.com/box-office-records/domestic/all-movies/cumulative/released-in-' + str(years))
        
        for page in pages:
            page = requests.get(URL + str(year) + '/' + str(page), headers=HEADERS)
            soup = bs(page.text.encode('latin1').decode('utf-8'), 'html.parser')
            tds = soup.find_all('td')

            title = []
            
            # Loop over td elements
            for td in tds:
                # If td element has b element
                if (td.b):
                    # if b anchor has text
                    if (td.b.text):
                        # Append the td to the list
                        title.append(td)
            
            for x in title:
                movies.append(x.text)
            print(movies)
    #print(movies)
    return movies
```

```python
#API call section 
url = "http://www.omdbapi.com/?apikey=" + "7524a562" + "&t="

responses21 = {'Title': [], 'Year': [], 'Rated': [], 'Released': [], 'Runtime': [], 'Genre': [], 'Director': [], 'Writer': [], 'Actors': [], 
'Plot': [], 'Language': [], 'Poster': [], 'Metascore': [], 'imdbRating': [], 'BoxOffice': [], 'Production': []}

i = 1
j = 1
for movie in scrape2021():
    movie_data = requests.get(url + movie + '&y=2021').json()
    #responses.append(movie_data)
    try:
        responses21['Title'].append(movie_data['Title'])
        responses21['Year'].append(movie_data['Year'])
        responses21['Rated'].append(movie_data['Rated'])
        responses21['Released'].append(movie_data['Released'])
        responses21['Runtime'].append(movie_data['Runtime'])
        responses21['Genre'].append(movie_data['Genre'])
        responses21['Director'].append(movie_data['Director'])
        responses21['Writer'].append(movie_data['Writer'])
        responses21['Actors'].append(movie_data['Actors'])
        responses21['Plot'].append(movie_data['Plot'])
        responses21['Language'].append(movie_data['Language'])
        responses21['Poster'].append(movie_data['Poster'])
        responses21['Metascore'].append(movie_data['Metascore'])
        responses21['imdbRating'].append(movie_data['imdbRating'])
        responses21['BoxOffice'].append(movie_data['BoxOffice'])
        responses21['Production'].append(movie_data['Production'])
        print('Good Data')

    except:
        print("Bad Data")
    pass    

    #delay for 65 seconds after every 100 calls    
    if i == 900:
        time.sleep(65)
        i = 0

#track number of call
        j += 1
    i += 1

else:
    print('----\nData Retrieval Complete\n----')
```

```python
#create dataframe from dictionary
movie_list_df_21 = pd.DataFrame.from_dict(responses21, orient='index')
movie_list_df_21 = movie_list_df_21.transpose()

#export to csv
movie_list_df_21.to_csv ('output/movie_list_df_21.csv', header=True)

#display preview
movie_list_df_21
```
