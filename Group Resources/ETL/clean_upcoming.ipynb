```python
import pandas as pd
import datetime as dt
from decimal import Decimal
```

```python
df = pd.read_csv('clean_upcoming_df_2022.csv')
df
```

```python
df['language'] = df['language'].fillna('NA')
df
```

```python
df['meta_score'] = df['meta_score'].fillna('0')
df
```

```python
df['release_date'].isnull().sum()
```

```python
df['release_date']=df['release_date'].fillna('1900-01-01')
df
```

```python
df['release_date'].isnull().sum()
```

```python
df['release_date'] = pd.to_datetime(df['release_date']).dt.date
df
```

```python
df['movie_rating']=df['movie_rating'].fillna('NA')
df
```

```python
#create movie table setup
movie = pd.DataFrame(df[['movie_id','title','movie_rating','genre_name','imdb_rating','director',
                        'writer','actors','box_office','run_time','release_date','meta_score']])
movie
```

```python
#export to csv
movie.to_csv ('final_movie_list_upcoming.csv', header=True, index=False)
movie
```

```python
#create movie table setup
movie2 = pd.DataFrame(movie[['movie_id','title','movie_rating', 'release_date', 'run_time', 'meta_score', 'imdb_rating','box_office']])

#export to csv
movie2.to_csv ('upcoming_movie.csv', header=True, index=False)

movie2
```

```python
#create movie table setup
genre = pd.DataFrame(movie[['movie_id','genre_name']]).rename(columns={'movie_id':'genre_id'})
#export to csv
genre.to_csv ('upcoming_genre.csv', header=True, index=False)

genre.head()
```

```python
#create movie table setup
movie_genre_juction = movie2.join(genre, how='outer')
movie_genre_juction = pd.DataFrame(movie_genre_juction[['movie_id','genre_id']])

#export to csv
genre.to_csv ('upcoming_movie_genre_juction.csv', header=True, index=False)

movie_genre_juction
```
