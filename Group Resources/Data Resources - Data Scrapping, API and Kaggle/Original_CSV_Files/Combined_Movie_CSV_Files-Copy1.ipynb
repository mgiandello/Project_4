```python
import pandas as pd
import datetime as dt
from decimal import Decimal
```

```python
load_df = pd.read_csv('output/Original_CSV_files/complete_movie_list_df_2021_2022.csv')
load_df = load_df.drop(['Unnamed: 0','Unnamed: 0.1','Production'], axis=1)
load_df
```

```python
load_df = load_df.fillna(0)
load_df
```

```python
load_df2=load_df['Runtime'].str.split(' ', expand=True)
load_df2
```

```python
load_df2 = load_df2.rename(columns={0: "a", 1: "c"})
load_df2
```

```python
# to append dataframe
load_df3 = load_df.join(load_df2)
load_df3
```

```python
load_df4=load_df3.drop(['c', 'Runtime'], axis=1)
load_df4
```

```python
load_df5 = load_df4.rename(columns={'a':'Runtime'})
load_df5
```

```python
#converting to proper types
load_df5['Released'] = pd.to_datetime(load_df['Released'], errors='coerce')
load_df5.Title = load_df.Title.astype('string')
load_df5.Rated = load_df.Rated.astype('string')
load_df5.Runtime = load_df.Runtime.astype('string')
#load_df['Runtime'] = load_df['Runtime'].str.split(' ', 1, expand=True)
load_df5.Genre = load_df.Genre.astype('string')
load_df5.Director = load_df.Director.astype('string')
load_df5.Writer = load_df.Writer.astype('string')
load_df5.Actors = load_df.Actors.astype('string')
load_df5.Director = load_df.Director.astype('string')
load_df5.Language = load_df.Language.astype('string')
load_df5['BoxOffice'] = load_df['BoxOffice'].replace('[\$,]', '', regex=True).astype(int)
load_df5.dtypes
```

```python
converted_df = load_df5
converted_df
```

```python
#export to csv
converted_df.to_csv ('output/Original_CSV_files/clean_movie_list_df_2021_2022.csv', header=True, index=False)
converted_df
```
