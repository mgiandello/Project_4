```python
def scrape():
    pages = np.arange(0,1,100)
    years = np.arange(2022,2023,1)
    URL = 'https://www.the-numbers.com/box-office-records/domestic/all-movies/cumulative/released-in-'
    HEADERS = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
    movies = []

    for year in years:
        #year = requests.get('https://www.the-numbers.com/box-office-records/domestic/all-movies/cumulative/released-in-' + str(years))
        
        for page in pages:
            page = requests.get(URL + str(year) + '/' + str(page), headers=HEADERS)
            soup = bs(page.text.encode('latin1').decode('utf-8'), 'html.parser')
            tds = soup.find_all('td')

            title = []
            
            # Loop over td elements
            for td in tds:
                # If td element has b element
                if (td.b):
                    # if b anchor has text
                    if (td.b.text):
                        # Append the td to the list
                        title.append(td)
            
            for x in title:
                movies.append(x.text)
            print(movies)
    #print(movies)
    return movies
```
