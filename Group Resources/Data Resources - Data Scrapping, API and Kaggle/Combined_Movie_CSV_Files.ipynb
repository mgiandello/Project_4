```python
movie_list2 = pd.read_csv('output/movies.csv')
movie_list2
```

```python
#clean dataframe
#movie_list2 = pd.DataFrame(movie_list2.rename(columns={'name':'title','rating':'movie_rating','genre':'genre_name','year':'year','released':'release_date',
#'score':'imdb_rating','votes':'votes', 'director':'Director', 'writer':'Writer', 'star':'Actors', 'country':'country', 'budget':'BoxOffice', 'gross':'gross',
#'company':'company','runtime':'runtime',}))
#movie_list2
```

```python
movie_list2 = movie_list2.fillna(0)
movie_list2
```

```python
release_date=movie_list2['released'].str.split('(', expand=True)
release_date
```

```python
release_date2 = release_date.rename(columns={0: "a", 1: "c"})
release_date2
```

```python
# to append dataframe
complete_movie_list_df = movie_list2.join(release_date2)
complete_movie_list_df
```

```python
complete_movie_list_df2=complete_movie_list_df.drop(['c', 'released'], axis=1)
complete_movie_list_df2
```

```python
complete_movie_list_df3 = complete_movie_list_df2.rename(columns={'a':'released','runtime':'Run'})
complete_movie_list_df3
```

```python
#converting to proper types
complete_movie_list_df3['released'] = pd.to_datetime(complete_movie_list_df3['released'], errors='coerce')
complete_movie_list_df3.name = complete_movie_list_df3.name.astype('string')
complete_movie_list_df3.rating = complete_movie_list_df3.rating.astype('string')
complete_movie_list_df3.Run = complete_movie_list_df3.Run.astype('string')
#load_df['Runtime'] = load_df['Runtime'].str.split(' ', 1, expand=True)
complete_movie_list_df3.genre = complete_movie_list_df3.genre.astype('string')
complete_movie_list_df3.writer = complete_movie_list_df3.writer.astype('string')
complete_movie_list_df3.star = complete_movie_list_df3.star.astype('string')
complete_movie_list_df3.director = complete_movie_list_df3.director.astype('string')
#complete_movie_list_df3['budget'] = complete_movie_list_df3['budget'].replace('[\$,]', '', regex=True).astype(int)
complete_movie_list_df3.dtypes
```

```python
complete_movie_list_df4 = complete_movie_list_df3.drop(['year','votes','country','gross','company'], axis=1)
complete_movie_list_df4
```

```python
#export to csv
complete_movie_list_df4.to_csv ('output/clean_movie_list_df2.csv', header=True, index=False)
complete_movie_list_df4
```
