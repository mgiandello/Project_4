```python
import pandas as pd
import datetime as dt
import csv
```

```python
movie_list_21 = pd.read_csv('output/movie_list_df_21.csv')
movie_list_21
```

```python
movie_list_22 = pd.read_csv('output/movie_list_df_2022.csv')
movie_list_22
```

```python
# to append movie list 2022 at the end of movie list 2021 dataframe
complete_movie_list_df_2021_2022 = movie_list_21.append(movie_list_22, ignore_index = True)
```

```python
complete_movie_list_df_2021_2022 = complete_movie_list_df_2021_2022.drop(['Plot','Poster'], axis=1)
complete_movie_list_df_2021_2022
```

```python
complete_movie_list_df_2021_2022 = complete_movie_list_df_2021_2022.fillna(0)
```

```python
#export to csv
complete_movie_list_df_2021_2022.to_csv ('output/complete_movie_list_df_2021_2022.csv', header=True)

#display preview
complete_movie_list_df_2021_2022
```
